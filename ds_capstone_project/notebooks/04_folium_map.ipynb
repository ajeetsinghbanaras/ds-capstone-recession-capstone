# 04 â€“ Interactive Map with Folium
Creates a simple map for the latest year using synthetic latitude/longitude coordinates.

```
import pandas as pd
import folium

df = pd.read_csv('../data/economic_indicators_clean.csv')
latest_year = df['year'].max()
latest = df[df['year'] == latest_year].copy()
coords = {
  'India': [20.59, 78.96],
  'USA': [37.09, -95.71],
  'Germany': [51.17, 10.45],
  'Japan': [36.20, 138.25],
  'Brazil': [-14.24, -51.93]
}
latest['lat'] = latest['country'].map(lambda c: coords.get(c, [0,0])[0])
latest['lon'] = latest['country'].map(lambda c: coords.get(c, [0,0])[1])
m = folium.Map(location=[20, 0], zoom_start=2)
for _, row in latest.iterrows():
    popup_text = f"{row['country']}<br>GDP: {row['gdp']:.0f}<br>Inflation: {row['inflation']:.1f}%"
    folium.CircleMarker(location=[row['lat'], row['lon']],
                        radius=8,
                        popup=popup_text,
                        color='blue',
                        fill=True,
                        fill_opacity=0.7).add_to(m)
m.save('folium_economic_map.html')
print('Map saved as folium_economic_map.html in notebooks folder')
```
